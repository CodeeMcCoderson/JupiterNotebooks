```python
# A Jointplot comprises three plots. Out of the three, one plot displays a bivariate graph which shows how the dependent variable(Y) 
# varies with the independent variable(X). Another plot is placed horizontally at the top of the bivariate graph and it shows 
# the distribution of the independent variable(X). see https://ajaytech.co/2020/09/28/8780/#:~:text=A%20Jointplot%20comprises%20three%20plots,the%20independent%20variable(X).
with sns.axes_style('white'):
  sns.jointplot("ins_premium","total", data=df,kind='hex') 
```

```python
with sns.axes_style('white'):
  sns.jointplot("ins_premium","total", data=df,kind='reg') 
```
