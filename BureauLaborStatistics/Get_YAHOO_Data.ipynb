```python
import datetime as dt
from pandas_datareader import data as pdr
import yfinance as yf
import pandas as pd
import numpy as np
```

```python
# Parameters
start = dt.datetime(2010,1,1)
end = dt.datetime.now()
dow = '^DJI'
sp500 = '^GSPC'
```

```python
# pull data back for specific stock, using Yahoo API, in specific date range
dow_df = yf.download(dow, start=start, end=end)
sp500_df = yf.download(sp500, start=start, end=end)
```

```python
dow_df['Date'] = dow_df.index
sp500_df['Date'] = dow_df.index
```

```python
dow_df[['Date']] = dow_df[['Date']].apply(pd.to_datetime)
sp500_df[['Date']] = dow_df[['Date']].apply(pd.to_datetime)
```

```python
dow_df.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis = 1, inplace = True)
```

```python
sp500_df.drop(['Open', 'High', 'Low', 'Adj Close', 'Volume'], axis = 1, inplace = True)
```

```python
dow_df['Month_Year'] = pd.to_datetime(dow_df['Date']).dt.to_period('M')
sp500_df['Month_Year'] = pd.to_datetime(sp500_df['Date']).dt.to_period('M')
dow_df['Day'] = pd.DatetimeIndex(dow_df['Date']).day
sp500_df['Day'] = pd.DatetimeIndex(sp500_df['Date']).day
```

```python
dow_df['Min_Date'] = (dow_df.groupby(dow_df['Month_Year'])
                          ['Day'].transform('min')
                       )
```

```python
dow_df['Max_Date'] = (dow_df.groupby(dow_df['Month_Year'])
                          ['Day'].transform('max')
                       )
```

```python
sp500_df['Min_Date'] = (sp500_df.groupby(sp500_df['Month_Year'])
                          ['Day'].transform('min')
                       )
```

```python
sp500_df['Max_Date'] = (sp500_df.groupby(sp500_df['Month_Year'])
                          ['Day'].transform('max')
                       )
```

```python
dow_df['Start_Month'] = np.where(dow_df['Day']== dow_df['Min_Date'], dow_df['Close'], False)
dow_df['End_Month'] = np.where(dow_df['Day']== dow_df['Max_Date'], dow_df['Close'], False)
```

```python
sp500_df['Start_Month'] = np.where(sp500_df['Day']== sp500_df['Min_Date'], sp500_df['Close'], False)
sp500_df['End_Month'] = np.where(sp500_df['Day']== sp500_df['Max_Date'], sp500_df['Close'], False)
```

```python
min_dow_df = dow_df[dow_df['Start_Month']>0]
```

```python
max_dow_df = dow_df[dow_df['End_Month']>0]
```

```python
min_sp500_df = sp500_df[sp500_df['Start_Month']>0]
```

```python
max_sp500_df = sp500_df[sp500_df['End_Month']>0]
```

```python
min_dow_df.drop(['Close', 'Date', 'Day', 'Min_Date', 'Max_Date', 'End_Month'], axis= 1, inplace = True)
max_dow_df.drop(['Close', 'Date', 'Day', 'Min_Date', 'Max_Date', 'Start_Month'], axis= 1, inplace = True)
dow_df = pd.merge(min_dow_df, max_dow_df, left_on = ['Month_Year'], right_on = ['Month_Year'])
```

```python
min_sp500_df.drop(['Close', 'Date', 'Day', 'Min_Date', 'Max_Date', 'End_Month'], axis= 1, inplace = True)
max_sp500_df.drop(['Close', 'Date', 'Day', 'Min_Date', 'Max_Date', 'Start_Month'], axis= 1, inplace = True)
sp500_df = pd.merge(min_sp500_df, max_sp500_df, left_on = ['Month_Year'], right_on = ['Month_Year'])
```

```python
dow_df['% Change'] = ((dow_df['End_Month'] - dow_df['Start_Month']) / dow_df['Start_Month'])* 100
```
